```python
#plotting the graphs
import pandas as pd
import matplotlib.pyplot as plot

#read data from csv
dataRead = pd.read_csv(r'C:\Users\Hansika\Downloads\NIC.csv')

#Make a series 
first = dataRead['Id']

#validate ids using regular expression
validateIds = first[first.str.match('^[0-9]{9}V$')== True]

#\Removing null values
nullRemoved = validateIds.dropna()

#Replace 'V' with a blank
vRemoved = nullRemoved.str.replace("V", ' ')

# make string version of vRemoved
stringIds = vRemoved.astype(str)

# make the new columns by splitting the stringVersion 
oneToTwoDigits = stringIds.str[0:2]#1-2 digits
twoToThreeDigits = stringIds.str[2:5]#3-5 digits
threeToFiveDigits = stringIds.str[5:8]#5-8 digits
lastDigit = stringIds.str[8:9]#last digit

#Scatter plot 1-2 digits vs Last Digit
plot.figure(figsize=(5,5))
x_data, y_data = (oneToTwoDigits, lastDigit)
plot.plot(x_data, y_data, 'ro',markersize=1)
plot.ylabel('Last  Digit')
plot.xlabel('1-2 digits')
plot.title("1-2 digits vs Last Digit")
plot.show()

#Scatter plot 3-5 digits vs Last Digit
plot.figure(figsize=(15,5))
x_data, y_data = (twoToThreeDigits, lastDigit)
plot.plot(x_data, y_data, 'ro', markersize=1)
plot.ylabel('Last  Digit')
plot.xlabel('3-5 digits')
plot.title("3-5 digits vs Last Digit")
plot.show()

#Scatter plot 5-8 digits vs Last Digit
plot.figure(figsize=(15,5))
x_data, y_data = (threeToFiveDigits, lastDigit)
plot.plot(x_data, y_data, 'ro',markersize=1)
plot.ylabel('Last Digit')
plot.xlabel('5-8 digits')
plot.title("5-8 digits vs Last Digit")
plot.show()


```
